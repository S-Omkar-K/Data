```python
#Aggregation on brand 
```

```python
autos["brand"]
```

```python
autos["brand"].value_counts()
```

```python
brand_price_avg = {}
brands = autos["brand"].unique()
for brand in brands:
    brand_price_avg[brand] = autos[autos["brand"] == brand]["usd_price"].mean().astype(int)
    
```

```python
brand_price_avg
```

```python
brand_mileage_avg = {}
brands = autos["brand"].unique()
for brand in brands:
    brand_mileage_avg[brand] = autos[autos["brand"] == brand]["odometer_km"].mean().astype(int)
    
```

```python
brand_mileage_avg
```

```python
#Formatting dates
```

```python
autos['date_crawled: time'] = autos['date_crawled'].str.split().str[1]
autos['date_crawled'] = (autos['date_crawled']
                         .str.replace('-', '')
                         .str.split().str[0].astype(int)
                        )

#there is no time in " ad_created", hence reformatting directly.

autos['ad_created'] = (autos['ad_created'].str.replace('-', '')
                       .str.split().str[0].astype(int)
                      )

autos['last_seen: time'] = autos['last_seen'].str.split().str[1]
autos['last_seen'] = (autos['last_seen']
                      .str.replace('-', '')
                      .str.split().str[0].astype(int)
                     )
autos[["date_crawled","ad_created","last_seen"]][0:5]
```
